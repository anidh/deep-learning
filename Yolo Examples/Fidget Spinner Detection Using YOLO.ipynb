```python
#importing all the libraries
import cv2
print("Open CV Version:"+str(cv2.__version__))
import matplotlib.pyplot as plt
%matplotlib inline
import time
import os
import numpy as np
from darkflow.net.build import TFNet
```

```python
options={"model":"cfg/tiny-yolo-voc-1c.cfg",
        "load":3900,
        "gpu":0.5,
        "threshold":0.1}
tfnet=TFNet(options)
```

```python
#Now defining our capture device
capture=cv2.VideoCapture(0)
colors=[tuple(255*np.random.rand(3)) for i in range(100)]
capture.set(cv2.CAP_PROP_FRAME_HEIGHT,1080)
capture.set(cv2.CAP_PROP_FRAME_WIDTH,1920)
```

```python
while(capture.isOpened()):
    startTime=time.time()
    ret,frame=capture.read()
    results=tfnet.return_predict(frame)
    if ret:
        for color,result in zip(colors,results):
            topLeft=(results[0]["topleft"]["x"],results[0]["topleft"]["y"])
            bottomRight=(results[0]["bottomright"]["x"],results[0]["bottomright"]["y"])
            label=results[0]["label"]
            confidence=results[0]["confidence"]
            frame=cv2.rectangle(frame,topLeft,bottomRight,(0,255,0),7)
            frame=cv2.putText(frame,label,topLeft,cv2.FONT_HERSHEY_COMPLEX,1,(0,0,0),2)
        cv2.imshow("frame",frame)
        print("FPS {:.1f}".format(1/(time.time()-startTime)))
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
    else:
        capture.release()
        cv2.destroyAllWindows()
        break
```
